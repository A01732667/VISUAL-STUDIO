```python
import pandas as pd
import numpy as np

df = pd.read_csv ('llamadas_filtradas.csv')
df.sample(10, random_state=42)
```

```python
import streamlit as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Preparaci√≥n de datos
df["Promesa_de_Pago"] = df["Promesa_de_Pago"].fillna(0)
df["Llamada_Contestada"] = df["Llamada_Contestada"].fillna(0)
df["Engagement"] = df["Llamada_Contestada"] + df["Promesa_de_Pago"]

# --- Tabla resumen por D√≠a y Rango Horario ---
st.subheader("üìã Tabla Resumen: D√≠a de la Semana y Rango Horario")

grouped = df.groupby(["Dia Semana", "Rango Hora"]).agg({
    "Promesa_de_Pago": "sum",
    "Llamada_Contestada": "sum",
    "ID Ally": "count"
}).rename(columns={"ID Ally": "Total Llamadas"}).reset_index()

grouped["Tasa Promesa"] = grouped["Promesa_de_Pago"] / grouped["Total Llamadas"]
grouped["Tasa Contestaci√≥n"] = grouped["Llamada_Contestada"] / grouped["Total Llamadas"]

st.dataframe(grouped.head(20))

# --- Heatmap ---
st.subheader("üå°Ô∏è Heatmap: Tasa de Promesa de Pago")

pivot_promesa = grouped.pivot(index="Dia Semana", columns="Rango Hora", values="Tasa Promesa")

fig1, ax1 = plt.subplots(figsize=(12, 6))
sns.heatmap(pivot_promesa, annot=True, fmt=".2f", cmap="YlGnBu", ax=ax1)
ax1.set_title("Tasa de Promesa de Pago por D√≠a de la Semana y Rango Horario")
ax1.set_xlabel("Rango Horario")
ax1.set_ylabel("D√≠a de la Semana")
plt.xticks(rotation=45)
st.pyplot(fig1)

# --- Gr√°fico de barras por Intensidad ---
st.subheader("üì∂ Tasa de Promesa de Pago por Intensidad de Llamada")

intensity_group = df.groupby("Intensidad").agg(
    promesas=("Promesa_de_Pago", "sum"),
    llamadas=("ID Ally", "count")
).reset_index()

intensity_group["Tasa Promesa"] = intensity_group["promesas"] / intensity_group["llamadas"]

fig2, ax2 = plt.subplots()
sns.barplot(data=intensity_group, x="Intensidad", y="Tasa Promesa", ax=ax2)
ax2.set_title("Tasa de Promesa de Pago por Intensidad de Llamada")
ax2.set_ylabel("Tasa de Promesa")
st.pyplot(fig2)
```
